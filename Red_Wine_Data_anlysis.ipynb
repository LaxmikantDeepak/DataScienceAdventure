```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import time
import random

random.seed(100)
```

```python
# File path to the csv files
data_red_wine =pd.read_csv("winequality-red.csv")
# set white as color
data_red_wine["color"] = "Blue"

# Print first 5 rows in the dataframe
data_red_wine.head()
```

```python
# describe the data
data_red_wine.describe(include="all")
```

```python
# inspect the variable types
data_red_wine.info()
```

```python
# check the correlation between the variables
# compute the correlation matrix
correlation_matrix = data_red_wine.corr()
figure = plt.figure(figsize = (14,12))
sns.heatmap(correlation_matrix, annot=True, cmap=plt.cm.Blues)
plt.show()
```

```python
# get some idea of the data
figure = plt.figure(figsize = (10,6))
sns.catplot(x="quality", y="fixed acidity", hue="color", kind="swarm", data=data_red_wine)
plt.title("Wine Quality as explained by Fixed Acidity")
```

```python
from sklearn.preprocessing import LabelEncoder
bins = (1, 6.5, 8)
group_names = ['bad', 'good']
data_red_wine['quality'] = pd.cut(data_red_wine['quality'], bins = bins, labels = group_names)
label_quality = LabelEncoder()
data_red_wine['quality'] = label_quality.fit_transform(data_red_wine['quality'])
data_red_wine['quality'].value_counts()
```
